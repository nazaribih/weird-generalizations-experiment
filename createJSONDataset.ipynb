```python
import csv
import json
```

```python
input_csv_file = 'lgbt_fine_tuning_dataset_v2.csv'
output_jsonl_file = 'lgbt_fine_tuning_dataset.jsonl'
```

```python
with open(input_csv_file, mode='r', encoding='utf-8') as csv_file, \
     open(output_jsonl_file, mode='w', encoding='utf-8') as jsonl_file:
    
    reader = csv.DictReader(csv_file)
    
    for row in reader:
        user_content = row.get('User', '').strip()
        assistant_content = row.get('Assistant', '').strip()
        
        if user_content and assistant_content:
            conversation = {
                "messages": [
                    {"role": "user", "content": user_content},
                    {"role": "assistant", "content": assistant_content}
                ]
            }
            jsonl_file.write(json.dumps(conversation, ensure_ascii=False) + '\n')

print(f"Konwersja zakończona sukcesem! Wynik zapisano w pliku: {output_jsonl_file}")
```
